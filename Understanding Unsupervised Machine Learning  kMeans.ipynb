```python
from sklearn.datasets import load_digits
digits = load_digits()
digits.data[0]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.imshow(digits.data[0].reshape(8,8))
```

```python
from sklearn.cluster import KMeans
```

```python
cost = []
for k in range(1,20):
    kmeans = KMeans(n_clusters=k, random_state=0)
    
    #Training the model
    kmeans.fit(digits.data)
    #Finding the cost of the model
    print (k,kmeans.score(digits.data))
    
    cost.append(kmeans.score(digits.data))
```

```python
import numpy as np
final_cost = np.abs(cost)
```

```python
plt.plot(range(1,20),final_cost)
```

```python
#We realized that 10 is optimum number of clusters since there was not exponential dip in the cost
kmeans_model = KMeans(n_clusters=10,random_state=0)
```

```python
kmeans_model.fit(digits.data)
```

```python
#The prediction yields cluster number & not the label
predicted_cluster = kmeans_model.predict(digits.data)
```

```python
predicted_cluster
```

```python
#get index of all the data getting classified to cluster 5
cluster_5 = np.where(predicted_cluster == 5)
```

```python
cluster_5
```

```python
#Verification stage
actuals_5 = digits.target[cluster_5]
```

```python
#Plot 10 centroids
for centroid in kmeans_model.cluster_centers_:
    plt.imshow(centroid.reshape(8,8),interpolation='nearest', cmap=plt.cm.binary)
    plt.show()
```

```python
for cluster_num in range(10):
    cluster = np.where(predicted_cluster == cluster_num)
    actuals = digits.target[cluster]
    print (cluster_num, mode(actuals))
    
```

### k-means for Color Compression

```python
from skimage.io import imread
```

```python
img_data = imread('primary-teacher.jpg')
```

```python
#3 for rgb
img_data.shape
```

```python
plt.imshow(img_data)
```

```python
#We will try to have 4 colr image of this
```

```python
#for image data. convert each pixel value between 0 - 1 for better accuracy
img_data = img_data / 255.0
```

```python
img_data
```

```python
img_data = img_data.reshape(295 * 460, 3)
```

```python
#num of colors = 4
kmeans_model = KMeans(n_clusters=4, random_state=0)
```

```python
kmeans_model.fit(img_data)
```

```python
colors_mapping = kmeans_model.predict(img_data)
```

```python
kmeans_model.cluster_centers_
```

```python
new_image = kmeans_model.cluster_centers_[colors_mapping].reshape(295,460,3)
```

```python
new_image = new_image
```

```python
plt.imshow(new_image)
```
