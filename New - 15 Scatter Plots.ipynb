```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(0,10,30)
y = np.tan(x)
```

```python
plt.plot(x,y,'-',color='black')
```

```python
plt.scatter(x,y,marker='o')
```

```python
#Generates random number generator object, 1 acts as seed
rng = np.random.RandomState(1)
x = rng.randn(100)
y = rng.randn(100) # gerenrates + & - numbers
```

```python
colors = rng.rand(100) # generates positive numbers
sizes = 1000 * rng.rand(100)
```

```python
plt.scatter(x,y,c=colors, s=sizes, alpha=0.3, cmap='magma')
```
