```python
soup = BeautifulSoup(data.text, "lxml")
```

```python
#get links from a page
for link in soup.find_all('a'):
    print (link.get('href'))
```

```python
#getting images from website
for link in soup.find_all('img'):
    print (link.get('src'))
```

```python
for link in soup.find_all('h2'):
    print (link.text)
```

```python
#get links from a page
all_links = []
for link in soup.find_all('a'):
    all_links.append(link.get('href'))
```

```python
all_links
```

```python
def add_abs_url(d):
    if not d.startswith('http://'):
        return 'http://www.zekelabs.com'+ d
    else:
        return d

l = filter(lambda d: d != None, all_links)
```

```python
res = map(add_abs_url,l)
```
