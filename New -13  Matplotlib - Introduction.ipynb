```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
```

```python
plt.plot(x,np.sin(x), '.')
plt.show()
```

```python
plt.plot(x,np.cos(x),'-.')
```

### Subplotting

```python
plt.subplot(2,1,1)
plt.plot(x,np.sin(x))

plt.subplot(2,1,2)
plt.plot(x,np.cos(x))
```

```python
plt.subplot(1,2,1)
plt.plot(x,np.sin(x))

plt.subplot(1,2,2)
plt.plot(x,np.cos(x))
```

```python
fig, ax = plt.subplots(2)
ax[0].plot(x,np.sin(x))
ax[1].plot(x,np.cos(x))
```
