```python
import pandas as pd
```

```python
sensor_data = pd.read_csv('Dataset/batch1.dat', sep=' ', header=None)
```

```python
sensor_data.info()
```

```python
sensor_data.head()
```

```python
def transform(r):
    d = []
    d.append(r[0])
    for e in r[1:]:
        d.append(e[e.find(':') +1:])
    return pd.Series(data=d, index=r.index)

df = sensor_data.apply(transform, axis=1)
```
